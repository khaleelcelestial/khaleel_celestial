```python
# Build a composition-based Order class that contains Address and PaymentInfo
class Address:
    def __init__(self, street, city, pincode):
        self.street = street
        self.city = city
        self.pincode = pincode

    def display(self):
        return f"{self.street}, {self.city} - {self.pincode}"
```

```python
class PaymentInfo:
    def __init__(self, method, amount):
        self.method = method
        self.amount = amount

    def display(self):
        return f"Payment Method: {self.method}, Amount: {self.amount}"
```

```python
class Order:
    def __init__(self, order_id, user, address, payment):
        self.order_id = order_id
        self.user = user              # object of User
        self.address = address        # object of Address
        self.payment = payment        # object of PaymentInfo

    def display_order(self):
        print(f"Order ID: {self.order_id}")
        print(f"User: {self.user.get_username()}")
        print(f"Shipping Address: {self.address.display()}")
        print(f"{self.payment.display()}")
```

```python
# Create Users
admin = AdminUser("admin1", "admin@mail.com", "admin123", 1)
customer = CustomerUser("cust1", "cust@mail.com", "cust123", "Bangalore")

# Display Profiles (Polymorphism)
admin.display_profile()
customer.display_profile()

# Create Composition Objects
addr = Address("MG Road", "Bangalore", "560001")
payment = PaymentInfo("Credit Card", 5000)

# Create Order
order = Order(101, customer, addr, payment)

# Display Order
order.display_order()
```

```python

#How to design user roles in an API using OOP
class Role:
    def __init__(self, name, permissions):
        self.name = name
        self.permissions = permissions
```

```python
class User:
    def __init__(self, username, role):
        self.username = username
        self.role = role

    def has_permission(self, permission):
        return permission in self.role.permissions
```

```python
admin_role = Role("admin", ["create", "delete", "update"])
user_role = Role("user", ["read"])

u1 = User("Alice", admin_role)
u2 = User("Bob", user_role)

print(u1.has_permission("delete"))  # True
print(u2.has_permission("delete"))  # False
```

```python
import time
import threading
import multiprocessing
```

```python
#Create a script that runs two IO-bound tasks using threads
def io_task(name):
    print(f"{name} started")
    time.sleep(2)  # Simulates waiting (API, DB, file)
    print(f"{name} finished")


def run_io_threads():
    print("\n--- IO-BOUND using THREADS ---")
    start = time.time()

    t1 = threading.Thread(target=io_task, args=("Task 1",))
    t2 = threading.Thread(target=io_task, args=("Task 2",))

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    print("Time:", time.time() - start)
run_io_threads()
```

```python
#Create a script that runs two CPU-bound tasks using multiprocessing
def cpu_task(n):
    total = 0
    for i in range(n):
        total += i * i
    return total


def run_cpu_multiprocessing():
    print("\n--- CPU-BOUND using MULTIPROCESSING ---")
    start = time.time()

    p1 = multiprocessing.Process(target=cpu_task, args=(10_000_000,))
    p2 = multiprocessing.Process(target=cpu_task, args=(10_000_000,))

    p1.start()
    p2.start()

    p1.join()
    p2.join()

    print("Time:", time.time() - start)
run_cpu_multiprocessing()

```

```python
#Demonstrate race condition with shared counter and fix it using a lock
counter = 0
lock = threading.Lock()


def increment_without_lock():
    global counter
    for _ in range(100000):
        counter += 1  # NOT SAFE
    return counter


def increment_with_lock():
    global counter
    for _ in range(100000):
        with lock:
            counter += 1  # SAFE
    return counter


def run_race_condition():
    global counter

    print("\n--- Race Condition (WITHOUT LOCK) ---")
    counter = 0

    t1 = threading.Thread(target=increment_without_lock)
    t2 = threading.Thread(target=increment_without_lock)

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    print("Expected:", 200000)
    print("Actual:", counter)

    print("\n--- Fixed with LOCK ---")
    counter = 0

    t1 = threading.Thread(target=increment_with_lock)
    t2 = threading.Thread(target=increment_with_lock)

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    print("Expected:", 200000)
    print("Actual:", counter)

run_race_condition()
```

```python
#Compare runtime for sequential vs threaded vs multiprocessing execution
def io_task(name):
    print(f"{name} started")
    time.sleep(2)  # Simulates waiting (API, DB, file)
    print(f"{name} finished")
    
def run_sequential():
    print("\n--- SEQUENTIAL EXECUTION ---")
    start = time.time()

    io_task("Task 1")
    io_task("Task 2")

    print("Time:", time.time() - start)


def run_threaded():
    print("\n--- THREADED EXECUTION ---")
    start = time.time()

    t1 = threading.Thread(target=io_task, args=("Task 1",))
    t2 = threading.Thread(target=io_task, args=("Task 2",))

    t1.start()
    t2.start()

    t1.join()
    t2.join()

    print("Time:", time.time() - start)


def run_multiprocess():
    print("\n--- MULTIPROCESS EXECUTION ---")
    start = time.time()

    p1 = multiprocessing.Process(target=io_task, args=("Task 1",))
    p2 = multiprocessing.Process(target=io_task, args=("Task 2",))

    p1.start()
    p2.start()

    p1.join()
    p2.join()

    print("Time:", time.time() - start)
run_sequential()
run_threaded()
run_multiprocess()
```

```python
#Convert a synchronous function to async
import time

def fetch_data():
    print("Start fetching...")
    time.sleep(2)  # blocking
    print("Data fetched")
```

```python
import asyncio

async def fetch_data():
    print("Start fetching...")
    await asyncio.sleep(2)  # non-blocking
    print("Data fetched")

# Run
asyncio.run(fetch_data())
```

```python
#Create multiple async functions using asyncio and Use asyncio.gather() to run multiple IO tasks concurrently
import asyncio

async def task1():
    print("Task 1 start")
    await asyncio.sleep(2)
    print("Task 1 end")

async def task2():
    print("Task 2 start")
    await asyncio.sleep(2)
    print("Task 2 end")

async def task3():
    print("Task 3 start")
    await asyncio.sleep(2)
    print("Task 3 end")

async def main():
    await asyncio.gather(
        task1(),
        task2(),
        task3()
    )

asyncio.run(main())
```

```python
#Write a small benchmark for sync vs async API-like calls
import time
import asyncio

# -----------------------
# SYNC VERSION
# -----------------------
def api_call_sync(name):
    print(f"{name} start")
    time.sleep(2)
    print(f"{name} end")

def run_sync():
    print("\n--- SYNC ---")
    start = time.time()

    api_call_sync("API 1")
    api_call_sync("API 2")
    api_call_sync("API 3")

    print("Sync Time:", round(time.time() - start, 2))


# -----------------------
# ASYNC VERSION
# -----------------------
async def api_call_async(name):
    print(f"{name} start")
    await asyncio.sleep(2)
    print(f"{name} end")

async def run_async():
    print("\n--- ASYNC ---")
    start = time.time()

    await asyncio.gather(
        api_call_async("API 1"),
        api_call_async("API 2"),
        api_call_async("API 3")
    )

    print("Async Time:", round(time.time() - start, 2))


# -----------------------
# MAIN
# -----------------------
if __name__ == "__main__":
    run_sync()
    asyncio.run(run_async())
```
